```python
def get_data_as_dataframe(
        limit : int = 50, 
        offset : int = 0, 
        select : str = "", 
        where  : str  = "",
        group_by : str = "",
        order_by : str = "",
        refine : dict = {},
        exclude : dict = {}
        )  -> pd.DataFrame:
    

    # query = f"?limit={limit}" \
    #         f"&offset={offset}" \
    query = ""
    if select != "":
        query += f"{'?' if len(query) == 0 else '&'}select={select}" 
    
    if where != "":
        query += f"{'?' if len(query) == 0 else '&'}where={where}" 

    if group_by != "":
        query += f"{'?' if len(query) == 0 else '&'}group_by={group_by}" 

    if order_by != "":
        query += f"{'?' if len(query) == 0 else '&'}order_by={order_by}" 

    if refine != {}:
        for key in refine:
            query += f"{'?' if len(query) == 0 else '&'}refine={key}%3A{refine[key]}"

    if exclude != {}:
        for key in exclude:
            query += f"{'?' if len(query) == 0 else '&'}exclude={key}%3A{exclude[key]}"
        
    if 0 < limit <= 100:
        query += f"{'?' if len(query) == 0 else '&'}limit={limit}" \
            f"&offset={offset}" 
        response = requests.get(base_datacount_url + query).json()
        total_count = response["total_count"] 
        data = response["results"]

    else:
        if limit == 0:
            response = requests.get(base_datacount_url + query+ f"{'?' if len(query) == 0 else '&'}limit=1").json()
            limit = int(response["total_count"])
        # le nombre de requete de limité à 100 entrés possible
        nbre_request = limit // 100
        # le nombre d'entré de la derniere requete
        last_limit = limit  % 100
        
        data = []
        for idx_req in range(nbre_request):
            limit_query = f"{'?' if len(query) == 0 else '&'}limit={100}" \
                    f"&offset={offset + idx_req*100}"
            response = requests.get(base_datacount_url + query + limit_query).json() 
            if "message" in response.keys():
                print(response)
                break
            data += response["results"]
        
        #envoie de la dernier requete
        limit_query = f"{'?' if len(query) == 0 else '&'}limit={last_limit}" \
                f"&offset={offset + (nbre_request)*100}"
        response = requests.get(base_datacount_url + query + limit_query).json() 
        data += response["results"]
        total_count = response["total_count"] 
    return total_count, pd.DataFrame(data )
```

## Fetching Data for a time frame



```python
from datetime import datetime
```

```python
def get_data_time_frame_as_dataframe(
        min_date : str, 
        max_date : str,
        limit : int = 0, 
        offset : int = 0, 
        select : str = "", 
        where  : str  = "",
        group_by : str = "",
        order_by : str = "",
        refine : dict = {},
        exclude : dict = {}):
    start_date = datetime.fromisoformat(min_date)
    end_date = datetime.fromisoformat(max_date)


    assert start_date <= end_date

    min_start_data = start_date.strftime("%Y-%m")


    _, df =   get_data_as_dataframe(limit=limit, refine={"date_j":f"{min_start_data}"})

    df['date'] = pd.to_datetime(df['date'])
    df = df.set_index('date')
    df = df.loc[min_date: max_date]
    return df

    
    
start_date = "2023-12-01"
end_date = "2023-12-03"

df = get_data_time_frame_as_dataframe(start_date, end_date, limit=0)
df
```
