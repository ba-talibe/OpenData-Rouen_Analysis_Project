## Données des comptage par heures

```python
counts_per_hour = tidf[["counts"]].groupby([pd.Grouper(freq="h")]).sum()
counts_per_hour
```

```python
counts_per_hour["date"] = counts_per_hour.index
counts_per_hour["day_name"] = counts_per_hour.date.dt.day_name()
counts_per_hour["hours"] = counts_per_hour.date.dt.hour
counts_per_hour
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


counts_per_hour['day_name'] = pd.Categorical(counts_per_hour['day_name'], categories=day_order, ordered=True)




heatmap_data = counts_per_hour.pivot_table(index='hours', columns='day_name', values='counts', aggfunc='sum', fill_value=0)


plt.figure(figsize=(19, 8))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='d', linewidths=.5)

# Customize plot labels and title
plt.title("Comptage hoaire pour chaque jour de la semaine sur toute l'etendue des données")
plt.xlabel("heures d'une journée")
plt.ylabel("Jours de la semaine")

# Show the plot
plt.show()
```
