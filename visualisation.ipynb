### Données des comptage par heures

```python
counts_per_hour = tidf[["counts"]].groupby([pd.Grouper(freq="h")]).sum()
counts_per_hour
```

```python
counts_per_hour["date"] = counts_per_hour.index
counts_per_hour["day_name"] = counts_per_hour.date.dt.day_name()
counts_per_hour["hours"] = counts_per_hour.date.dt.hour
counts_per_hour
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


counts_per_hour['day_name'] = pd.Categorical(counts_per_hour['day_name'], categories=day_order, ordered=True)




heatmap_data = counts_per_hour.pivot_table(index='hours', columns='day_name', values='counts', aggfunc='sum', fill_value=0)


plt.figure(figsize=(19, 8))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='d', linewidths=.5)

# Customize plot labels and title
plt.title("Comptage hoaire pour chaque jour de la semaine sur toute l'etendue des données")
plt.xlabel("heures d'une journée")
plt.ylabel("Jours de la semaine")

# Show the plot
plt.show()
```

### Diagrammmme de chaleur de la dernier semaine 

```python
from datetime import datetime, timedelta

end_date = tidf.index[0]
start_date = end_date - timedelta(days=7)
weekly_date_range = pd.date_range(start=start_date, end=end_date, freq="H")

start_date, end_date
```

```python
counts_average_per_hour = tidf[["counts"]].loc[weekly_date_range].groupby([pd.Grouper(freq="h")]).mean()



counts_average_per_hour["date"] = counts_average_per_hour.index
counts_average_per_hour["day_name"] = counts_average_per_hour.date.dt.day_name()
counts_average_per_hour["hours"] = counts_average_per_hour.date.dt.hour


day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


counts_average_per_hour['day_name'] = pd.Categorical(counts_average_per_hour['day_name'], categories=day_order, ordered=True)




heatmap_data = counts_average_per_hour.pivot_table(index='hours', columns='day_name', values='counts', aggfunc='mean', fill_value=0)


plt.figure(figsize=(19, 8))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".0f", linewidths=.5)

# Customize plot labels and title
plt.title("Carte de chaleur de pour chaque jour de la semaine derniere")
plt.xlabel("Heures d'une journée")
plt.ylabel("Jours de la semaine")

# Show the plot
plt.show()
```

### Diagramme de chaleur du dernier mois

```python
from datetime import datetime, timedelta

end_date = tidf.index[0]
start_date = end_date - timedelta(days=365)
monthly_date_range = pd.date_range(start=start_date, end=end_date, freq="H")

start_date, end_date
```

```python
monthly_date_range = monthly_date_range.intersection(tidf.index)
daily_counts_average = tidf[["counts"]].loc[monthly_date_range].groupby([pd.Grouper(freq="h")]).mean()



daily_counts_average["date"] = daily_counts_average.index
daily_counts_average['month_name'] = daily_counts_average['date'].dt.strftime('%B')
daily_counts_average["hours"] = daily_counts_average.date.dt.hour


month_order = [
    'January', 'February', 'March', 'April', 'May', 'June', 'July',
    'August', 'September', 'October', 'November', 'December'
]
daily_counts_average['month_name'] = pd.Categorical(daily_counts_average['month_name'], categories=month_order, ordered=True)




heatmap_data = daily_counts_average.pivot_table(index='hours', columns='month_name', values='counts', aggfunc='mean', fill_value=0)


plt.figure(figsize=(19, 8))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".0f", linewidths=.5)

# Customize plot labels and title
plt.title("Carte de chaleur de la moyenne horaire pour chaque mois de la dernier anneé")
plt.xlabel("heures d'une journée")
plt.ylabel("Jours de la semaine")

# Show the plot
plt.show()
```
